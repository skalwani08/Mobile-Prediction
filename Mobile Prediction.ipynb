# Creating X and Y

```python
x=mobile_proj.values[:,0:-1]
y=mobile_proj.values[:,-1]
```

```python
x.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x=scaler.fit_transform(x)
#scaler.fit(x)
#x=scaler.transform(x)
print(x)
```

# Building The Model

```python
from sklearn.model_selection import train_test_split

#split the data into and train

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#predicting using the Kneighbors_classifier

from sklearn.neighbors import KNeighborsClassifier
model_KNN=KNeighborsClassifier(n_neighbors=int(np.sqrt(len(x_train))),metric='manhattan')

#fit the  model on the data and predict the values
model_KNN.fit(x_train,y_train)

y_pred=model_KNN.predict(x_test)
print(list(zip(y_test,y_pred)))
```

```python
import numpy as np
np.sqrt(len(x_train))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(y_test,y_pred)
print(cfm)
print()

print("Classification report: ")

print(classification_report(y_test,y_pred))

acc=accuracy_score(y_test, y_pred)
print("Accuracy of the model: ",acc)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(y_test,y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(y_test,y_pred))

acc=accuracy_score(y_test, y_pred)
print("Accuracy of the model: ",acc)
```

# Logistic Regression

```python
#Recreating X and Y
x=mobile_proj.values[:,0:-1]
y=mobile_proj.values[:,-1]
```

```python
from sklearn.model_selection import train_test_split

#split the data into test and train
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
#create a model object
classifier=LogisticRegression(multi_class='multinomial')
#train the model object
classifier.fit(x_train,y_train)# fit function is used to train the model

y_pred=classifier.predict(x_test)
print(list(zip(y_test,y_pred)))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(y_test,y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(y_test,y_pred))

acc=accuracy_score(y_test, y_pred)
print("Accuracy of the model: ",acc)
```

#### Insight: So overall we can conclude that logistic regression works well when compared to KNN model
